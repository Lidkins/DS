Напишіть функції, що:
* рохраховує мінімальне та максимальне значення
* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```python
import random
#генеруэмо тисячу чисел в діапазоні від 0 до 10
list_with_random_numbers = [random.randint(0,10) for i in range(0,1000)]
data = list(list_with_random_numbers)
# шукаю мінімум та максимум
minimum = min(data)
print(minimum)
maximum = max(data)
print(maximum)
# розрахунок середнього значення
s = sum (data)
n = len(data)
mean = s / n
print (mean)
# розрахунок медіани
if len(data) % 2 == 0:
    left_index = int(len(data) / 2) - 1 
    right_index = int(len(data) / 2)
    med = (data[left_index] + data[right_index]) / 2
else:
    inx = len(data) // 2
    med = data[inx]
    
print(med)
# розрахунок частоти кожного слова в послідовності
import collections
counter = collections.Counter(data)
print(counter)
# розрахунок середнього квадратичного відхилення
from math import sqrt
for i in range (len(data)):
    m = (i - mean)**2
    m += 1
sigma_data = sqrt(m / (n-1))
standart_error_data = sigma_data/sqrt(n)
print(standart_error_data)
```
