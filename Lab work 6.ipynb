```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import seaborn as sns
```

```python
np.random.seed(0)

df = pd.DataFrame(data={'a':np.random.randint(0, 100, 30),
                        'b':np.random.randint(0, 100, 30),
                        'c':np.random.randint(0, 100, 30)})
df.head()
```

Створити візуалізацію, аналогічно рисунку 
 - перші 3 графіки візуалізувати, викорстовуючи значення із df
 - останній це пряма пропорційність

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,  2, figsize=(15,8))
ax1.plot(df.a)
ax2.plot(df.b)
ax3.plot(df.c)
ax4.plot(df.b, df.b)
fig.tight_layout( )
```

Створити візуалізацію, аналогічно рисунку 

```python
fig, ax = plt.subplots(figsize=(15,8))
ax.plot(df.a)
```

Створити візуалізацію, аналогічно рисунку 
- використовуйте колонки `a` та `b`

```python
fig, (ax1, ax2) = plt.subplots(2, figsize=(15,8))
ax1.plot(df.a)
ax2.plot(df.b)
```

Створити візуалізацію, аналогічно рисунку 
- використовуйте колонки `a` та `b`

```python
fig, ax = plt.subplots(figsize=(15,8))
line1, = ax.plot(df.a)
line2, = ax.plot(df.b)
```

Створити візуалізацію, аналогічно рисунку 
- використовуйте колонки `a` та `b`
- задайте стиль 'darkgrid' за допомогою команди `sns.set_style`

```python
fig, ax = plt.subplots(figsize=(15,8))
line1, = ax.plot(df.a)
line2, = ax.plot(df.b)
sns.set_style("darkgrid")
```

Створити візуалізацію, аналогічно рисунку 
- для колонки `a` використайте червоний колір та лінію формату `-.`

- для колонки `b` використайте помаранчевий колір та товщину `10`

- для колонки `c` використайте жовтий колір та товщину `1` і маркер `o`


```python
fig, ax = plt.subplots(figsize=(15,8))
line1, = ax.plot(df.a, color='red', linestyle='dashdot')
line2, = ax.plot(df.b, color='orange', linewidth=10)
line3, = ax.plot(df.c, color='yellow', marker='o', linestyle='dashed', linewidth=1, markersize=5)
sns.set_style("darkgrid")
```

Створити візуалізацію, аналогічно рисунку 
- і не забудьте про легенду :)

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, figsize=(15,8))
ax1.plot(df.a)
ax1.legend(['Line A'], loc='lower right',  shadow=True)
ax2.plot(df.b)
ax2.legend(['Line B'], loc='center left',  shadow=True)
ax3.plot(df.c)
ax3.legend(['Line C'], loc='upper right',  shadow=True)
sns.set_style("darkgrid")
```

Створити візуалізацію, аналогічно рисунку 
- використайте томатний колір та відстань між стовбцями 0.5

```python
fig, ax = plt.subplots(figsize=(20,8))
ax.bar(x = df.index, height = df.a, color='tomato', align='center', width=0.5)
ax.legend(['a'], loc='upper right',  shadow=True)
```

Створити візуалізацію, аналогічно рисунку 
- добавте всі підписи та правильний маркер

```python
fig, ax = plt.subplots(figsize=(20,10))
ax.plot(df.a, marker='^', linewidth=2, markersize=7)
ax.set_xlabel('This is the X axis', fontsize=14)
ax.set_ylabel('This is the Y axis', fontsize=14)
ax.set_title('This is the title',fontsize=16)
```
